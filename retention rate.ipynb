## Calculate the retention rate for upcoming 7 days (n = [0,7]) after the install on daily bases. Retention is calculated by dividing the count of the users who opened the app after n days from install. For instance, on March 12, 2020, 1,000 users have installed the app. On March 13 400 of them has returned/reopened. Thus, the Day 1(D1) will be 400 / 1000 = 40%.

```python
unique_installs = users.groupby("install_date",as_index=False)['id'].count().rename(columns={"id":"installs"})
unique_events = events.groupby("event_date",as_index=False)['id'].nunique().rename(columns={"id":"events"})
```

```python
unique_installs['install_date'] = pd.to_datetime(unique_installs['install_date'])
```

```python
retention_df = merged_df.groupby(["event_date","install_date"],as_index=False)['id'].nunique().rename(columns={'id':'returned_users'})

retention_df = retention_df[(retention_df['event_date'] - retention_df['install_date']).dt.days <=7]

retention_df = unique_installs.merge(retention_df,on='install_date')

retention_df['retention_rate'] = (retention_df['returned_users']/retention_df['installs'] * 100).round(2)

retention_df.sort_values(['install_date','event_date'],inplace=True)

retention_df = retention_df[['install_date','event_date','installs','returned_users','retention_rate']]
```

```python
retention_df
```

## For each day calculate the count of churned users. Churned are called the users who came the day before but hasn’t returned the next day. For instance, on March 10 1,000 users opened the app. On March 11 1,200 from which 400 were the new installed ones. Therefore, the churned will be 1200 – 400 – 1000 = -200. In other words, the churned users are equal the active users of the given day minus the new installs minus the active users of the previous day.

```python
day_install_day_open = retention_df[retention_df['event_date'] == retention_df['install_date']]\
            [['install_date','event_date','returned_users','installs']]
```

```python
unique_events['event_date'] = pd.to_datetime(unique_events['event_date'])
```

```python
churn_df = unique_events.merge(day_install_day_open,on='event_date').drop("install_date",axis=1)\
            .rename(columns={"events":"total_events","returned_users":"day_install_day_open"})


churn_df['yesterday_opens'] = churn_df['total_events'].shift(1)

churn_df.dropna(inplace=True)

churn_df.drop("installs",inplace=True,axis=1)

churn_df['churn_rate_number'] = churn_df['total_events'] - churn_df['day_install_day_open'] - churn_df['yesterday_opens']

churn_df['churn_rate'] = (churn_df['churn_rate_number']/churn_df['yesterday_opens']).round(2) * 100
```

## Calculate the conversion rate from open to purchase for each day. This could be done by dividing the unique users count who did a purchase by the unique users count who opened the app.

```python
purchase_events = merged_df[(merged_df['event_name'] == 'purchase')]
purchase_events
```

```python
open_events
```

```python
open_users_count = open_events.groupby('event_date')['id'].nunique()
purchase_users_count = purchase_events.groupby('event_date')['id'].nunique()

conversion_rate_df = pd.merge(open_users_count, purchase_users_count, left_index=True, right_index=True, how='outer')
conversion_rate_df = conversion_rate_df.rename(columns={'id_x': 'Open_users', 'id_y': 'Purchase_users'})

conversion_rate_df['Conversion_rate'] = (conversion_rate_df['Purchase_users'] / conversion_rate_df['Open_users'])

conversion_rate_df
```
